# Data Visualization 

Data Visualization is the graphical representation of information and data using visual elements like charts, graphs, maps, and dashboards. It helps people understand complex data sets by presenting them in a more accessible and easy-to-digest visual format.

## Why Data Visualization is Important
* **Simplifies Complex Data** - It turns large, complex datasets into visual summaries that are easier to understand.
* **Improves Data Interpretation** - Helps identify patterns, trends, and outliers quickly that might go unnoticed in raw data.
* **Facilitates Better Decision-Making** - Decision-makers can grasp insights quickly and act on data-driven conclusions.
* **Enhances Communication** Makes it easier to share insights with stakeholders through visual storytelling.
* **Saves Time** Users can comprehend large amounts of information at a glance rather than reading reports or spreadsheets.
* **Supports Exploratory Data Analysis (EDA)** In data science and machine learning, itâ€™s critical for understanding the structure and relationships in the data.
* **Engages Audience** Visually appealing graphics attract more attention and help in better engagement compared to text-heavy reports.

### Important Libraries in Python 
* **Pandas**: A powerful Python library used for data manipulation and analysis with DataFrame support.
* **Matplotlib**: A foundational plotting library for creating static, customizable 2D charts in Python.
* **Seaborn**: A Python library built on Matplotlib that provides high-level, attractive statistical visualizations.
* **Plotly (Python)**: An interactive and animated plotting library that enables dynamic, web-based visualizations.

* Line inside the box: The median (Q2).
* Whiskers: Extend to show the range (excluding outliers).
* Dots outside whiskers: Outliers.

## Plotting With Plotly 
[Plotting with Plotly](https://plotly.com/python/) <br>
Plotly is a powerful library for creating interactive and animated plots. It provides a wide range of visualization options, including 3D plots, geographical plots, and interactive charts that allow for zooming, panning, and hovering over elements for additional information.

Key Concepts to Remember
* **Interactive Visualizations**: Plotly allows for interactive elements like zoom, pan, hover, and more.
* **Express vs Graph Objects**:
  * plotly.express is a simpler interface for quick plotting.
  * plotly.graph_objects provides more fine-grained control for complex visualizations.
* **Figure Objects**: In Plotly, everything is encapsulated in a Figure object, which can be customized and updated.
* **Web-based**: Plots can be rendered in Jupyter Notebooks or exported to HTML, making them suitable for web apps.



**Additional Notes on Plotly**
* **Hover Info**: Plotly automatically displays additional information (e.g., values) on hover.
* **Interactivity**: You can zoom, pan, and filter data interactively.
* **Exporting**: You can export the plots to an HTML file for sharing or embedding into websites using fig.write_html("plot.html").

```python
import plotly.express as px 
```

```python
df=sns.load_dataset('penguins')
```

```python
df.sample()
```

### 13. Bubble Chart 
A bubble chart is an extension of a scatter plot that displays three dimensions of data: the x and y positions represent two variables, while the size of each bubble represents a third variable, helping visualize relationships and compare magnitude simultaneously.

```python

df_clean = df.dropna(subset=["bill_length_mm", "bill_depth_mm", "body_mass_g", "species"])

# Create Bubble Chart
fig = px.scatter(
    df_clean,
    x="bill_length_mm",
    y="bill_depth_mm",
    size="body_mass_g",            # Must be numeric with no NaNs
    color="species",
    title="Bubble Chart: Bill Dimensions vs Body Mass (Penguins)",
    labels={
        "bill_length_mm": "Bill Length (mm)",
        "bill_depth_mm": "Bill Depth (mm)",
        "body_mass_g": "Body Mass (g)"
    }
)
fig.show()
```

### 14. Bar Chart
A bar chart is a type of chart that displays data as rectangular bars. Each bar represents a value.

```python
px.bar(df,x='species',y='body_mass_g', title='Bar Chart: Species vs Body Mass (Penguins)',color='species',labels={'species':'Species','body_mass_g':'Body Mass (g)'})
```

### 15. 3d Line Plot
Shows a connected path in 3D space â€” useful for trends or sequences.


```python

df = sns.load_dataset("penguins").dropna()

fig = px.line_3d(
    df.sort_values(by="body_mass_g"),  # sort helps visualize direction
    x="bill_length_mm",
    y="bill_depth_mm",
    z="body_mass_g",
    color="species",
    title="3D Line Plot: Bill Dimensions and Body Mass"
)
fig.show()
```

### 16. Animated Bar Plot: Avg Body Mass per Species Over Islands


```python
grouped = df.groupby(['island', 'species'])['body_mass_g'].mean().reset_index()

# Animated bar chart
fig = px.bar(
    grouped,
    x="species",
    y="body_mass_g",
    color="species",
    animation_frame="island",
    title="Average Body Mass by Species Across Islands",
    labels={"body_mass_g": "Avg Body Mass (g)"}
)
fig.show()
```

### 17. Animated Bubble Chart 

```python
df["frame"] = df.groupby("species").cumcount()

fig = px.scatter(
    df,
    x="flipper_length_mm",
    y="body_mass_g",
    size="bill_length_mm",
    color="species",
    animation_frame="frame",
    hover_name="island",
    title="Animated Bubble Plot: Flipper Length vs Body Mass",
    range_x=[170, 240],
    range_y=[2500, 6500]
)
fig.show()
```
