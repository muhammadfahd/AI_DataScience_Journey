# Plotting with Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.<br>
For Detail [check out](https://seaborn.pydata.org/examples/index.html)

ðŸŒŸ Unique Features of Seaborn
* **Built-in Themes & Styles** : Beautiful default themes and color palettes make charts visually appealing out of the box.
* **Integration with Pandas** : Directly works with DataFrames and column names, making it very intuitive.
* **Statistical Plotting** : High-level functions like sns.boxplot(), sns.violinplot(), sns.regplot(), and sns.lmplot() automatically compute statistical estimations like confidence intervals and regression lines.
* **Categorical Data Handling**: Seaborn shines with categorical plots like catplot(), stripplot(), swarmplot(), and more.
* **Multi-Plot Grids**: Functions like pairplot(), facetgrid(), and jointplot() allow for complex visualizations across multiple variables with minimal code.
* **Color Palettes** : Provides diverse color palette options (deep, muted, pastel, dark, colorblind, etc.), useful for both aesthetic and accessibility

## Box Plot

```python
# we can specifiy figure size
plt.figure(figsize=(13,6))

#changing the scattr point colors
palette={'First':'blue','Second':'red','Third':'green'}
sns.boxplot(data=df, x='sex',y='fare', hue='class', palette=palette)


plt.xlabel('Sex')
plt.ylabel('fare')
plt.title('Box plot with Seaborn')

# saving a plot 
# multiple methods to save a plot  300dpi, dots per inch
plt.savefig('./output_plots/box_plot.png', dpi=300)  # save as PNG

```

## Creating MultiPlots
In this section, we will learn how to create multiplots. Multiplots are useful when we want to display multiple plots in a single figure. 

### Method 1: Using FacetGrid â€“ Create a grid of plots by class and sex

```python
g = sns.FacetGrid(df, col="class", row="sex")
g.map_dataframe(sns.histplot, x="age", kde=True)
plt.show()
```

### Method 2: Using catplot â€“ Show survival count per class

```python
sns.catplot(data=df, x="class", hue="survived", kind="count", col="sex")
plt.show()
```

### Method 3: Using pairplot â€“ Relationship between multiple numeric features
Visualize relationships and separability between classes (e.g., survived vs. not).

```python
df_clean = df.dropna(subset=["age", "fare", "pclass", "survived"])  # Clean NaNs
sns.pairplot(df_clean, hue="survived", vars=["age", "fare", "pclass"])
plt.show()
```

### Method 4: Using lmplot â€“ Regression lines grouped by hue

```python
sns.lmplot(data=df, x="age", y="fare", hue="survived", col="class", height=4)
plt.show()
```
